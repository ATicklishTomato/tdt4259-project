```python
# Create a map with a point for each cluster
m = folium.Map(location=[40.75, -73.9], zoom_start=11)
for cluster in clusters:
    # Add a point for each cluster
    folium.Marker(
        [cluster[0], cluster[1]],
        tooltip=cluster[2],
        popup=cluster[2]
    ).add_to(m)
    # Add cluster number as text
    folium.map.Marker(
        [cluster[0], cluster[1]],
        icon=folium.DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            # HTML code for the text in red
            html='<div style="font-size: 12pt; color: black; font-weight: bold;">%d</div>' % cluster[2]
        )
    ).add_to(m)
m
```
